```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

# Régression linéaire simple

```python

# Données
poids = np.array([90, 50, 48, 60, 120, 100, 110, 80, 85, 75, 90, 70, 65]).reshape(-1, 1)  # Variable X (poids en KG)
taille = np.array([180, 160, 161, 155, 200, 180, 195, 170, 180, 175, 205, 180, 165])  # Variable y (taille en cm)
```

```python

# Création et entraînement du modèle de régression linéaire
model = LinearRegression()
model.fit(poids, taille)
```

```python
# Coefficient et intercept du modèle
print(f"Coefficient (pente): {model.coef_[0]:.3f}")
print(f"Intercept: {model.intercept_:.2f}")
```

```python

# Prédiction des valeurs de taille à partir du poids
taille_pred = model.predict(poids)

# Affichage des résultats
print("Coefficient de régression (pente) :", model.coef_[0])
print("Intercept :", model.intercept_)
print("Erreur quadratique moyenne (MSE) :", mean_squared_error(taille, taille_pred))
print("Score R² :", r2_score(taille, taille_pred))

# Visualisation des données et de la droite de régression
plt.scatter(poids, taille, color="blue", label="Données réelles")
plt.plot(poids, taille_pred, color="red", label="Ligne de régression")
plt.xlabel("Poids (KG)")
plt.ylabel("Taille (cm)")
plt.title("Régression linéaire : Taille en fonction du Poids")
plt.legend()
plt.show()
```

# Régression linéaire multiple

```python
age = np.array([25, 18, 21, 30, 45, 35, 40, 28, 32, 26, 50, 29, 22])
```

```python
X = np.column_stack((poids, age))  # Matrice de deux colonnes : poids et âge
Y = taille

# Créer et ajuster le modèle de régression linéaire multiple
model = LinearRegression()
model.fit(X, Y)
```

```python
print(f"Coefficients (pentes): Poids = {model.coef_[0]:.2f}, Âge = {model.coef_[1]:.2f}")
print(f"Intercept: {model.intercept_:.2f}")
print("Erreur quadratique moyenne (MSE) :", mean_squared_error(taille, taille_pred))
print("Score R² :", r2_score(taille, taille_pred))
```

```python
taille_pred = model.predict(X)

# Visualisation en 3D des données et de la surface de régression
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(poids, age, taille, color='blue', label='Données')
ax.set_xlabel("Poids (KG)")
ax.set_ylabel("Âge (ans)")
ax.set_zlabel("Taille (cm)")

# Tracer la surface de régression
poids_range = np.linspace(poids.min(), poids.max(), 10)
age_range = np.linspace(age.min(), age.max(), 10)
Poids, Age = np.meshgrid(poids_range, age_range)
Taille_pred = model.intercept_ + model.coef_[0] * Poids + model.coef_[1] * Age
ax.plot_surface(Poids, Age, Taille_pred, color='red', alpha=0.5)

plt.title("Régression linéaire multiple : Poids, Âge vs Taille")
plt.legend()
plt.show()
```

# Régression logistique

ous allons utiliser LogisticRegression de scikit-learn pour créer un modèle de régression logistique et prédire si une personne est en surpoids en fonction de son poids et de son âge.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
poids = np.array([90, 50, 48, 60, 120, 100, 110, 80, 85, 75, 90, 70, 65])
age = np.array([25, 18, 21, 30, 45, 35, 40, 28, 32, 26, 50, 29, 22])
surcharge = np.array([1, 0, 0, 0, 1, 0, 1, 0, 0, 0, 1, 0, 0])
```

```python
X = np.column_stack((poids, age))  # Matrice de deux colonnes : poids et âge
Y = surcharge
```

```python
model = LogisticRegression()
model.fit(X, Y)
```

```python
predictions = model.predict(X)
probas = model.predict_proba(X)
```

```python
print("Matrice de confusion :\n", confusion_matrix(Y, predictions))
print("\nRapport de classification :\n", classification_report(Y, predictions))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(poids, age, surcharge, color='blue', label='Données réelles')
# Tracer la surface de décision de la régression logistique
poids_range = np.linspace(poids.min(), poids.max(), 10)
age_range = np.linspace(age.min(), age.max(), 10)
Poids, Age = np.meshgrid(poids_range, age_range)
probas_surface = model.predict_proba(np.c_[Poids.ravel(), Age.ravel()])[:, 1]
probas_surface = probas_surface.reshape(Poids.shape)
ax.plot_surface(Poids, Age, probas_surface, color='red', alpha=0.3)
ax.set_xlabel("Poids (KG)")
ax.set_ylabel("Âge (ans)")
ax.set_zlabel("Probabilité de surcharge")
plt.title("Régression logistique : Poids, Âge vs Probabilité de surcharge")
plt.show()
```
